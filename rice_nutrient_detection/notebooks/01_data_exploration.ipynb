```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python

BASE_DIR = "../data/rice_plant_lacks_nutrients"
CLASS_DIRS = {
	"Nitrogen": "Nitrogen(N)",
	"Phosphorus": "Phosphorus(P)",
	"Potassium": "Potassium(K)"
}
```

```python
print("Dataset structure:")
for cls, d in CLASS_DIRS.items():
	class_path = os.path.join(BASE_DIR, d)
	count = len(os.listdir(class_path)) if os.path.exists(class_path) else 0
	print(f"{cls}: {count} images")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for i, (cls, d) in enumerate(CLASS_DIRS.items()):
	class_path = os.path.join(BASE_DIR, d)
	files = [f for f in os.listdir(class_path) if not f.startswith(".")]
	if files:
		img = Image.open(os.path.join(class_path, files[0])).convert("RGB")
		axes[i].imshow(img)
		axes[i].set_title(cls)
		axes[i].axis("off")
plt.tight_layout()
plt.show()
```
